```python
import numpy as np                 # For numerical operations
import pandas as pd                # For data manipulation and analysis
import matplotlib.pyplot as plt    # For plotting graphs
import seaborn as sns              # For data visualization (based on matplotlib)


import json

%matplotlib inline
```

```python
fname = '../data/results.json'
def get_required_data(filename):
    filtered_db = []
    with open(filename, 'r') as file:
        db = json.load(file)
    for entry in db:
        if entry["_source"]['content_metadata']['language'] == 'en':
            filtered_db.append(entry)
    return filtered_db

db = get_required_data(fname)
```

```python
seen = []
filtered_db = []
for article in db:
    if article["_source"]['content_metadata']['context']['body_en'] not in seen:
        seen.append(article["_source"]['content_metadata']['context']['body_en'])
        filtered_db.append(article)
len(filtered_db)
    

    
```

```python
with open("../data/cleaned_articles.json", "w") as fout:
    json.dump(filtered_db, fout, indent=2, ensure_ascii=False)
```
